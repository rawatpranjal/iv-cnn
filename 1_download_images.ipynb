```python
import os
import shutil
from tqdm import tqdm
from PIL import Image
import torch
from torchvision import models, transforms
import numpy as np

# Path to the main dataset directory
main_dataset_path = '/content/drive/MyDrive/dataset2'  # Set the path to your dataset here

# Initialize and set the ResNet model to evaluation mode
model = models.resnet50(pretrained=True)
model.eval()

# Define the image processing transformations
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # Resize the image to 64x64
    transforms.Grayscale(num_output_channels=1),  # Convert the image to grayscale
    transforms.ToTensor(),
])

# Relevant ImageNet classes for cars
car_classes = [511, 609, 661, 751, 817]  # Convertible, Jeep/Landrover, Model T, Racer, Sports Car

# Function to process a folder and find the image with the highest probability of being a car
def process_folder(folder_path):
    max_prob = 0
    selected_image_tensor = None
    selected_embedding = None

    if os.path.exists(folder_path) and os.listdir(folder_path):
        for image_file in os.listdir(folder_path):
            image_path = os.path.join(folder_path, image_file)
            try:
                img = Image.open(image_path).convert('RGB')
                img_t = transform(img)
                batch_t = torch.unsqueeze(img_t, 0)

                with torch.no_grad():
                    out = model(batch_t)
                probabilities = torch.nn.functional.softmax(out, dim=1)
                car_prob = probabilities[0, car_classes].sum()

                if car_prob > max_prob:
                    max_prob = car_prob
                    selected_image_tensor = img_t.flatten().numpy()  # Save flattened pixel tensor
                    selected_embedding = out.flatten().numpy()  # Save the embedding

            except Exception as e:
                print(f"Error processing image {image_path}: {e}")

    return selected_image_tensor, selected_embedding

# Process all folders and collect tensors
pixel_tensors = []
embedding_tensors = []
for folder_name in tqdm(os.listdir(main_dataset_path), desc="Processing Folders"):
    folder_path = os.path.join(main_dataset_path, folder_name)
    image_tensor, embedding = process_folder(folder_path)
    if image_tensor is not None and embedding is not None:
        pixel_tensors.append(image_tensor)
        embedding_tensors.append(embedding)

# Convert lists to numpy arrays
pixel_tensors = np.stack(pixel_tensors) if pixel_tensors else np.array([])
embedding_tensors = np.stack(embedding_tensors) if embedding_tensors else np.array([])

# Save pixel tensors to a file
pixel_tensors_path = '/content/drive/MyDrive/pixel_tensors_2.pt'  # Specify the path and file name
torch.save(pixel_tensors, pixel_tensors_path)

# Save embedding tensors to a file
embedding_tensors_path = '/content/drive/MyDrive/embedding_tensors_2.pt'  # Specify the path and file name
torch.save(embedding_tensors, embedding_tensors_path)

```

### Save to tensor

```python
import os
from PIL import Image
import torch
from torchvision import transforms
import numpy as np
from tqdm import tqdm

# Path to the main dataset directory
main_dataset_path = '/content/drive/MyDrive/iv-cnn/dataset'  # Replace with the path to your dataset

# Define the image processing transformations with grayscale conversion
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),  # Convert image to grayscale
    transforms.Resize((640, 360)),  # Resize the image to 1280x720
    transforms.ToTensor(),  # Convert the image to a PyTorch tensor
])

# Function to process a folder and pick the largest image in quality
def process_folder(folder_path):
    if os.path.exists(folder_path) and os.listdir(folder_path):
        largest_image_file = None
        largest_size = 0

        for image_file in os.listdir(folder_path):
            image_path = os.path.join(folder_path, image_file)
            if os.path.isdir(image_path):
                continue  # Skip directories
            try:
                size = os.path.getsize(image_path)
                if size > largest_size and Image.open(image_path).size >= (640, 360):
                    largest_size = size
                    largest_image_file = image_file
            except Exception as e:
                print(f"Error checking image {image_path}: {e}")

        if largest_image_file:
            image_path = os.path.join(folder_path, largest_image_file)
            try:
                img = Image.open(image_path).convert('RGB')
                img_t = transform(img)
                return img_t
            except Exception as e:
                print(f"Error processing image {image_path}: {e}")

    return None

# Process all folders and collect pixel tensors and queries
pixel_tensors = []
queries = []
for folder_name in tqdm(os.listdir(main_dataset_path), desc="Processing Folders"):
    folder_path = os.path.join(main_dataset_path, folder_name)
    image_tensor = process_folder(folder_path)
    if image_tensor is not None:
        pixel_tensors.append(image_tensor)
        queries.append(folder_name)  # Save the query (folder name)

# Convert list of tensors to a single tensor
pixel_tensors = torch.stack(pixel_tensors) if pixel_tensors else torch.tensor([])

# Save pixel tensors and queries to files
pixel_tensors_path = '/content/drive/MyDrive/iv-cnn/pixel_tensors.pt'  # Specify the path and file name for pixel tensors
torch.save(pixel_tensors, pixel_tensors_path)

queries_path = '/content/drive/MyDrive/iv-cnn/queries_rich.npy'  # Specify the path and file name for queries
np.save(queries_path, queries)
```
